```python
import pandas as pd

# Cargar datos desde la URL
url = 'https://raw.githubusercontent.com/Sinchiguano/chicagoUniversity/main/cta_Ridership.csv'
df = pd.read_csv(url)

# Convertir la columna de fecha a tipo DateTime
df['date'] = pd.to_datetime(df['date'])

# Configurar la columna de fecha como el índice
df.set_index('date', inplace=True)

# Filtrar los datos para la estación 'Garfield-South Elevated'
Garfield_South_Elevated = df[df['stationname'] == 'Garfield-South Elevated']

# Mostrar las primeras filas del DataFrame filtrado para verificar
print(Garfield_South_Elevated.head())

# Agrupar los datos por fecha y obtener estadísticas
# Por ejemplo, contar el número de viajes por fecha para la estación 'Garfield-South Elevated'
viajes_Garfield_South_Elevated = Garfield_South_Elevated.resample('D').size()

# Mostrar las primeras filas del resultado
print(viajes_Garfield_South_Elevated.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(11, 4)})

# Asumiendo que 'Garfield_South_Elevated' es el DataFrame filtrado para esa estación
Garfield_South_Elevated['rides'].plot(linewidth=0.5)
plt.title('Número de viajes en la estación Garfield-South Elevated a lo largo del tiempo')
plt.xlabel('Fecha')
plt.ylabel('Número de viajes')
plt.show()

```
